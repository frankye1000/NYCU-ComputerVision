```python
imgLeft,  imgLeft_gray,  imgLeft_rgb  = read_img('test/m1.jpg')
imgRight, imgRight_gray, imgRight_rgb = read_img('test/m2.jpg')
```

```python
plt.imshow(imgLeft_gray, cmap='gray')
plt.show()
plt.imshow(imgRight_gray, cmap='gray')
plt.show()
```

```python
def SIFT(img):
    SIFT_Detector = cv2.SIFT_create()
    kp, des = SIFT_Detector.detectAndCompute(img, None)

    return kp, des

def plot_sift(gray, rgb, kp):
    tmp = rgb.copy()
    img = cv2.drawKeypoints(image=gray, 
                            keypoints=kp, 
                            outImage=tmp, 
                            color=(255,0,0), 
                            flags=cv2.DRAW_MATCHES_FLAGS_DRAW_OVER_OUTIMG)
    return img
```

```python
kpLeft, desLeft   = SIFT(imgLeft_gray)
kpRight, desRight = SIFT(imgRight_gray)

kp_left_img  = plot_sift(imgLeft_gray, imgLeft_rgb, kpLeft)
kp_right_img = plot_sift(imgRight_gray, imgRight_rgb, kpRight)
total_kp = np.concatenate((kp_left_img, kp_right_img), axis=1)
plt.imshow(total_kp)
```
